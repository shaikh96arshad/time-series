```python
import pandas as pd
import pandas_datareader as pdr
from datetime import datetime
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
```

```python
df_tesla.head(3)
```

```python
df_tesla['Open'].plot()
```

```python
df_tesla['10Daysroll'] = df_tesla['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df_tesla[['Open','10Daysroll']].plot(xlim=['2020-01-01','2022-12-12'],figsize=(15,4))
```

```python
df_tesla['30DaysRoll'] = df_tesla['Open'].rolling(window=30,min_periods=1).mean()
df_tesla['50DaysRoll'] = df_tesla['Open'].rolling(window=50,min_periods=1).mean()
```

```python
df_tesla[['Open','10Daysroll','30DaysRoll','50DaysRoll']].plot(xlim=['2020-01-01','2022-12-31'],figsize=(15,4))
```

# Expanding or Cumulitive Moving Average

```python

df_tesla['Open'].expanding().mean().plot(figsize=(15,4))
```

# Extended Moving Average

```python
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1,adjust=False).mean()
df_tesla['EMA_0.3'] = df_tesla['Open'].ewm(alpha=0.3,adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1','EMA_0.3']].plot(figsize=(15,4))
```

# Exponential weight moving average

```python
df_tesla['EMA5Days'] = df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open','EMA_0.3','EMA5Days']].plot(figsize=(15,4))
```
